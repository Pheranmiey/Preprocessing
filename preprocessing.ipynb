```python
df.drop(columns = ['record_id'], inplace = True)
df
```

```python
# filling the missing values in df['fuel_unit'] with 'missing'
df['fuel_unit'].fillna('missing', inplace = True )
```

```python
df.isnull().sum()
```

```python
# checking for numerical and categorical variables , 'O' refers to object and it basically means categorical variables, while '!O' refers to numerical variables 
numerical_features = [feature for feature in df.columns if df[feature].dtypes !='O']
print('number of numerical variables: ', len(numerical_features))
df[numerical_features].head()
```

```python
categorical_features = [feature for feature in df.columns if df[feature].dtypes =='O']
print('number of categorical variables: ', len(categorical_features))
df[categorical_features].head()
```

```python
# categorical values need to be changed to numerical values, that is how they can be worked on
```

```python
categorical_features
```

```python
# handling categorical values
categorical_features = ['plant_name_ferc1', 'fuel_type_code_pudl', 'fuel_unit']
for variables in categorical_features:
    print (np.unique(df[variables]))
```

```python
df = pd.get_dummies(df, columns=categorical_features)
df
# note the new columns in df are the sum total of all the unique variables in categorical features
```

```python
# getting all the features in our data frame (df) and storing them in feature_scale_df
```

```python
feature_scale_df = [feature for feature in df]
```

```python
df[feature_scale_df]
```

```python
# Now we need to scale our data to be in same range, i used minmax scaler here,there are other scalers like robust scaler,standard scaler etc, you can check them out
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[feature_scale_df])
new_df = pd.DataFrame(scaler.transform(df), columns=[feature_scale_df])
```
